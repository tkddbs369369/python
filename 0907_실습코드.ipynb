##**14181**
흔한 수학 문제 중 하나는 주어진 점이 어느 사분면에 속하는지 알아내는 것이다. 사분면은 아래 그림처럼 1부터 4까지 번호를 갖는다. "Quadrant n"은 "제n사분면"이라는 뜻이다.



예를 들어, 좌표가 (12, 5)인 점 A는 x좌표와 y좌표가 모두 양수이므로 제1사분면에 속한다. 점 B는 x좌표가 음수이고 y좌표가 양수이므로 제2사분면에 속한다.

점의 좌표를 입력받아 그 점이 어느 사분면에 속하는지 알아내는 프로그램을 작성하시오. 단, x좌표와 y좌표는 모두 양수나 음수라고 가정한다.

```python
x = int(input())
y = int(input())
global x_
global y_
if x > 0:
  x_ = True
else:
  x_ = False

if y > 0:
  y_ = True
else:
  y_ = False


if x_ == True and y_ == True:
  print(1)
elif x_ == False and y_ == False:
  print(3)
elif x_ == True and y_ == False:
  print(4)
else:
  print(2)
```

## 9498
- 시험 점수를 입력받아 90 ~ 100점은 A, 80 ~ 89점은 B, 70 ~ 79점은 C, 60 ~ 69점은 D, 나머지 점수는 F를 출력하는 프로그램을 작성하시오.

```python
score = int(input())

if score >= 90:
  print('A')
elif score >= 80:
  print('B')
elif score >= 70:
  print('C')
elif score >= 60:
  print('D')
else:
  print('F')
```

- 두 정수 A와 B를 입력받은 다음, A+B를 출력하는 프로그램을 작성하시오.



















```python
################################     10950 for로 풀기
a = int(input())
for a in range(a):
    c, b = input().split()
    c = int(c)
    b = int(b)

    print(c + b)

```

```python
################################     10950 while로 풀기
a = int(input())
count = 0
while count < a:
    count = count + 1
    c, b = map(int, input().split())
    c = int(c)
    b = int(b)

    print(c + b)
```
